```python
import matplotlib.pyplot as plt
import numpy as np

# Energy content data (in MJ/kg)
fuels = ['Algae Biodiesel', 'Gasoline', 'Diesel', 'Coal']
energy_content = [38.6, 46.4, 45.5, 24.0]

# Create a bar plot
plt.figure(figsize=(8, 6))
x = np.arange(len(fuels))
plt.bar(x, energy_content)
plt.xticks(x, fuels)
plt.ylabel('Energy Content (MJ/kg)')
plt.title('Comparison of Energy Content: Algae Biodiesel vs Fossil Fuels')

# Rotate x-axis labels
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```
