```python
import numpy as np
import sympy as sp
def trap(a,b,n,f):
    h= (b-a)/n
    Integral= (1/2)*(f.subs(x,a) + f.subs(x,b))
    for i in range(1,n):
        Integral=Integral + f.subs(x,a+(i*h))
    return sp.N(h*Integral)

def simp(a,b,n,f):
    if n%2!=0:
        raise ValueError("not even")
    h=(b-a)/n
    Integral=f.subs(x,a)+f.subs(x,b)
    for i in range(1,n):
        if i%2==0:
            Integral= Integral+ 2*f.subs(x, a+(i*h))
        else:
            Integral= Integral+ 4*f.subs(x, a+(i*h))
    return (h/3)*Integral
x=sp.symbols('x')
f=sp.sin(x)
a=0
b=np.pi
errortr=[]
n=4
l=trap(a,b,n,f)
errortr.append(abs(2-l))
print(l)
n=8
l=trap(a,b,n,f)
errortr.append(abs(2-l))
print(l)
n=16
l=trap(a,b,n,f)
errortr.append(abs(2-l))
print(l)


    
print(errortr)
        
        
```

```python
errorsi=[]
n=4
s=simp(a,b,n,f)
errorsi.append(abs(2-s))
print(s)
n=8
s=simp(a,b,n,f)
errorsi.append(abs(2-s))
print(s)
n=16
s=simp(a,b,n,f)
errorsi.append(abs(2-s))
print(s)

print(errorsi)
```

```python
import matplotlib.pyplot as plt
plt.plot([4, 8, 16], errorsi, label="Simpson's Error", marker='o')
plt.plot([4, 8, 16], errortr, label="Trapezoidal Error", marker='o')
plt.xlabel('n (Number of intervals)')
plt.ylabel('Error')
plt.title("Error Comparison: Simpson's vs Trapezoidal")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#The error approaches 0 as we increase the number of subintervals in both the methods
#Simpsons method is better for this as it captures the curvature of the function more accurately than the linear approximation used 
#in the Trapezoidal rule.
```
