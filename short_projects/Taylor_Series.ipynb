```python
import numpy as np
import matplotlib.pyplot as plt
import math
import sympy as sp
```

```python
def Ts(var,order,func,point):
    ts=0
    for i in range (order+1):
        y=func.diff(var,i).subs(var,point)*((var-point)**i)*(1/(math.factorial(i)))
        ts=ts+y
    return ts
```
