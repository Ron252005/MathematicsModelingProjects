### Reason behind Fluctuation converging to 0. 

####  Law of large number
As the number of trials approaches a very large number then the average obtained from the trials resembles closely to the actual value obtained from probability calculation or mean. This characteristic reduces the fluctuations seen with small trials, as given below: 

STANDARD Deviation formula:
                             SD = ([(Σ(x(i) - m)^2] ) /N)^(1/2)
                             
When N increases to a large number the Deviation or Fluctuation is significantly less as the actual value starts converging as N is indirectly proportional to SD in the above formula, as Fluctuation is (SD)^2.

## Question 1 (C)

```python
import numpy as np
import matplotlib.pyplot as plt

t = 0
h = 0
fractionProb1 = 0  # this is the fraction from previous point initialized at zero
dict  = {}

fig, axs = plt.subplots(2, figsize=(8, 12))

# ___________________________________Plot for 1,000_________________________________________

for i in range(1000):
    result = np.random.choice(['H', 'T'])
    if result == 'H':
        h += 1
    t += 1
    
    fractionProb = h / t
    instFluc = fractionProb - fractionProb1
    fractionProb1 = fractionProb
    
    dict[t] = instFluc

axs[0].plot(list(dict.keys()), list(dict.values()), color='red')
axs[0].set_title('Fluctuations Graph for 1,000')
axs[0].set_xlabel('Total Cases')
axs[0].set_ylabel('Instantaneous Fluctuations')
axs[0].grid(True)



# ___________________________________Plot for 10,000_________________________________________

t = 0
h = 0
fractionProb1 = 0  # this is the fraction from previous point initialized at zero
dict  = {}

for i in range(10000):
    result = np.random.choice(['H', 'T'])
    if result == 'H':
        h += 1
    t += 1    
    fractionProb = h / t
    instFluc = fractionProb - fractionProb1
    fractionProb1 = fractionProb
    
    dict[t] = instFluc

axs[1].plot(list(dict.keys()), list(dict.values()), color='blue')
axs[1].set_title('Fluctuations Graph for 1,000')
axs[1].set_xlabel('Total Cases')
axs[1].set_ylabel('Instantaneous Fluctuations')
axs[1].grid(True)
```

If India emerges victorious, my friend owes me Rs. 100; conversely, if India loses, I owe my friend Rs. 1000. The total investment amounts to Rs. 1100. In the event of India's defeat, my willingness to pay from the total Rs. 1100 suggests a subjective belief in the likelihood of India winning, estimated at 10/11. This implies a high level of confidence, in the probability of India's victory, while the probability of India losing is deemed very low. Consequently, this subjective assessment indicates that the odds are in favor of India. This is a condition of Subjective probability.


### Below is just a unique representation of the subjective probability with numbers and not a prooven method as its subjective.
For any event, we know,

                                        P(E)  = 1 - P(not E)                 -(eq 1)
                        
If the Event E here is India winning then we can use this equation in our problem.

Also rationally and ideally with assumptions probability of india winning or losing both should be 0.5.
If someone is ready to bet 100 on the win and ready to loose 1000 if it doesn't win. Then both conditions should have equal probability values.  
#### This disparity in SUBJECTIVE PROBABILITY arises due to personal bias, experience, belief in the team and many other parameters.
                
                                      100*P(E) = 1000*P(not E)                  -(eq 2)

Putting value from equation 1 in equation 2;
                100 - 100*P(not E) = 1000*P(not E)
             => 1- P(not E) = 10 * P(not E)
             => 1 = 11 * P(not E)
             => P(not E) = 1/11
From equation 1,
        
        P(E) = 1 - P(not E)
        P(E) = 10 / 11

#### P(E) = 10/11
#### Probability percent of India winning : (10/11) *100 = 0.909 *100 =  90.9 %
