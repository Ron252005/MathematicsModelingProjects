```python
import numpy as np
import matplotlib.pyplot as plt

def getSteps(): 
    transitions = np.array([[6,2], [0, 0], [0, 4], [0, 0], [5, 6], [6,5], [5,8], [0, 0]])  
    #Possible transitions for the snake

    current = 0
    steps = 0
    
    while current != 8:
        transition = np.random.randint(2)
        current = transitions[current, transition]
        steps += 1
    return steps

b = np.zeros(1000)
for i in range(1000):
    plt.plot(i,np.sum(b)/i, '.')
    b[i] = getSteps()

print("Average steps taken to reach the end:", np.mean(b))
```
